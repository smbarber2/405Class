```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('covid_confirmed_usafacts.csv')

STATE = df[df['State'] == 'OH'] #Filters out dataframe to only Ohio

CovidWeek = STATE.loc[:, ['County Name'] + STATE.columns[-7:].tolist()] #Only uses the last week

focused = CovidWeek.melt(id_vars=['County Name'], var_name='Date', value_name='Cases') ##Focuses down the information needed and customizes it down to something more useable
focused['Date'] = pd.to_datetime(focused['Date'])  #Date in the right format? It looks weird without it.
focused.head()

#focusedCounty = focused.iloc[:, ['

#print(focused)  
#This wouldn't work! it doesn't show it in a good format. Therefore, I found how to use plt. Its really hard. This is hard lol

for county, data in focused.groupby('County Name'):   #This has the data shown within the graph being the seperate counties
    plt.plot(data['Date'], data['Cases'], marker='.')  #This sorts the counties with Date as the X, Cases as the Y

plt.rcParams["figure.figsize"] = (10,5)   #Adjusts the size of the graph so the numbers are legible

plt.show()

```

As you can see, cases do trend up slightly in this final week of 2020! I wish it could be more detailed, but you can clearly see how the line gradually gets higher as marked by the points.
