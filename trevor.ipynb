```python
import pandas as pd

df = pd.read_csv("ACSST5Y2020.S1501-Data.csv", low_memory=False)
covid = pd.read_csv("super_covid_data.csv")
covid
```

```python
##Filter to 2020 covid data only
data2020 = covid.filter(like='2020')
covid2020 = pd.concat([covid.iloc[:, :5], data2020], axis=1)
covid2020
```

```python
##Focus on California
cacovid = covid2020[covid2020['State'] == 'CA']
cacovid
```

```python
##Check on the last week of 2020 and add a row to sum all cases each day
cases = cacovid.filter(like='cases')
finalweek = cases.iloc[:, -7:]
finalweek.loc['totals'] = finalweek.sum()
finalweek
```

```python
##As the total number of cases increased in greater volume each day, we can assume cases in California were increasing the last week of 2020.
```

```python
##Only taking useful columns from the census sheet
df = df[['NAME', 'S1501_C01_001E', 'S1501_C01_002E', 'S1501_C01_003E', 'S1501_C01_004E', 'S1501_C01_005E', 'S1501_C01_006E', 'S1501_C01_007E', 'S1501_C01_008E', 'S1501_C01_009E', 'S1501_C01_010E', 'S1501_C01_011E', 'S1501_C01_012E']]

##Reorganizing the sheet to fit with super_covid
update = df.drop(0)
update = update.rename(columns={'NAME': 'County Name','S1501_C01_001E': '18-24 - Total polled for educational attainment', 'S1501_C01_002E': '18-24 - less than high school education', 'S1501_C01_003E': '18-24 - High school grad', 'S1501_C01_004E': "18-24 - Some college or associate's", 'S1501_C01_005E': "18-24 - Bachelor's degree",'S1501_C01_006E': '25 and up - Total polled for educational attainment','S1501_C01_007E': '25 and up - less than 9th grade.', 'S1501_C01_008E': '25 and up - between 9th and 12th grade', 'S1501_C01_009E': '25 and up - high school grad', 'S1501_C01_010E': '25 and up - Some college, no degree', 'S1501_C01_011E': "25 and up - Associate's degree", 'S1501_C01_012E': "25 and up - Bachelor's degree"}) 

##The county name column was originally populated by "county name, state abbreviation", so we remove all except "county name" for merging purposes
update['County Name'] = update['County Name'].str.split(',').str[0]

update
```

```python
##Merging modified educational attainment data with the 2020 covid sheet, as the educational attainment census is for the year 2020

merged = pd.merge(update, covid2020, on='County Name')
merged
```
